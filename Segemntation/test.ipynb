```python
from skimage import io, transform, img_as_float 
```

```python
def int_to_float(f):
    image = io.imread(f)   # [0-255]
    image = img_as_float(image)  #[0-1]
    image = (image - image.min()) * (1 / (image.max() - image.min()))  #
    image = transform.resize(image, (256, 256))  #
    return image
```

```python
path = "/cis/phd/yz4891/PaddleSeg/dataset/annotations/*.png"  #.jpg
collections = io.ImageCollection(path, load_func=int_to_float) 
#ImageCollection int_to_float
```

```python
path1 = "/cis/phd/yz4891/PaddleSeg/dataset/annotations/"
for i in range(len(collections)):
    io.imsave(path1+file+'.png', collections[i])  #
```
